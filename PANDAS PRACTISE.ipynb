## Question 13

```python
#What are the top 5 most common last name in the dataset? 
#Bonus: Can you figure this out in one line of pandas code? (For simplicity treat the a title such as MD as a last name, 
#for example Caroline Conley MD can be said to have the last name MD)

most_name = hb['name'].str.split().str[-1].value_counts().nlargest(5) #.str is used to keep converting to strings. 
#split() is to seperate the two words by comma once a space is sin. 
#-1 reads background and slides sinces name as been divided into two different words as a result of its space. 
#value_counts are used to count the number of values

most_name
```

```python
#Another method is to create a function that can be used for everything

def last_name(x):
    return x.split()[-1]
```

```python
#then apply the last_name

hb['name'].apply(last_name).value_counts()[:5]
```
